```python
""" Import libraries """
import matplotlib.pyplot as plt
import xarray as xr
import cartopy
```

```python
""" Load the dataset """

input_file = "air.2m.gauss.2021.nc"

# download the data file if needed
import os
if not os.path.exists(input_file):
    url = f"https://psl.noaa.gov/thredds/fileServer/Datasets/ncep.reanalysis2/gaussian_grid/air.2m.gauss.2021.nc"
    ! curl --output {input_file} {url}

# open the dataset using xarray
temp_ds = xr.open_dataset(input_file, chunks = -1)

""" Calculate the hottest temperature recorded in each season. """

# xarray is built on top of pandas, so we can use groupby for this
season_groups = temp_ds.groupby('time.season')

# calculate the max in each season
season_max_temp = season_groups.max()

# extract temperature (and also use 'squeeze' to remove the pesky single-item level dimension)
max_temp_xr = season_max_temp['air'].squeeze()

# pull out coordinates
seasons = max_temp_xr.season
lat = max_temp_xr.lat
lon = max_temp_xr.lon

# force the calculation
max_temp_xr.load();
```

# Cartopy

For mapping, we'll use `cartopy`.  Cartopy defines *projection* as the map projection on which the dataset will be drawn.  It defines the *transform* of the dataset as the projection on which the dataset exists.  They can be the same, but they don't have to be.

Let's just explore projections.

```python

# set the coordinates of bloomington, IN
bloomington_lon = -86.526386
bloomington_lat = 39.165325

projection = cartopy.crs.Orthographic(central_longitude = bloomington_lon, central_latitude= bloomington_lat)
transform = cartopy.crs.PlateCarree()

# create a figure
fig, ax = plt.subplots(subplot_kw = dict(projection = projection))

# draw continent boundaries
ax.coastlines()

# plot a point corresponding to Bloomington, IN
ax.plot(bloomington_lon, bloomington_lat, 'r*', transform = transform)

# set a global extent
ax.set_global()

plt.show()
```

```python
""" Draw data on a map. """
# set the coordinates of bloomington, IN
bloomington_lon = -86.526386
bloomington_lat = 39.165325

projection = cartopy.crs.Orthographic(central_longitude = bloomington_lon, central_latitude= bloomington_lat)
transform = cartopy.crs.PlateCarree()

# create a figure
fig, ax = plt.subplots(subplot_kw = dict(projection = projection))

# draw the DJF data
max_temp_djf = max_temp_xr.sel(season = 'DJF')
lat = max_temp_djf.lat
lon = max_temp_djf.lon
ax.pcolormesh(lon, lat, max_temp_djf, transform = transform)

# draw continent boundaries
ax.coastlines()

# plot a point corresponding to Bloomington, IN
ax.plot(bloomington_lon, bloomington_lat, 'r*', transform = transform)

# set a global extent
ax.set_global()

plt.show()
```

```python

""" Make the plot look nice. """

# set the lat/lon of Portland, OR
portland_lon = -122.6765
portland_lat = 45.5231

# set the plot projection
projection = cartopy.crs.Orthographic(central_longitude = portland_lon, central_latitude= portland_lat)
# set the data transform
transform = cartopy.crs.PlateCarree()

# set the style
plt.style.use('seaborn-v0_8-poster')

# set the colormap
cmap = cmocean.cm.thermal


# generate a 2x2 panel figure with a better aspect ratio
# (have all subpanels use the same map projection)
fig, axs2d = plt.subplots(
    2,2,
    figsize = (8, 6),
    sharex = True,
    sharey = True,
    subplot_kw=dict(projection = projection),
    )

# reduce horizontal spacing between the plots
#plt.subplots_adjust(wspace = -0.5)

# create a vector of plots (first index is the top left)
axs = axs2d.ravel()

# set panel labels
labels = ['(a)', '(b)', '(c)', '(d)']

for ax, season, label in zip(axs, seasons, labels):
    # select the data for this season and convert to C
    temp_this_season = max_temp_xr.sel(season = season) - 273.15


    # plot temperature using pcolormesh (each value in the array is assigned a color)
    cplt = ax.pcolormesh(
        lon,
        lat,
        temp_this_season,
        cmap = cmap,
        vmin = -30,
        vmax = 50,
        transform = transform,
        )

    # set the title to the season
    ax.set_title(str(season.values))

    # add a panel label in the lower left of each panel
    ax.text(0.0, 0.0, label, transform = ax.transAxes, fontsize = 16, color = 'black' )

    # add coastlines
    ax.coastlines()

    # add US states
    ax.add_feature(cartopy.feature.STATES)

# add a star over portland in the first panel
axs[0].plot(portland_lon, portland_lat, 'r*', transform = transform)


# add a common colorbar
fig.colorbar(cplt, ax = axs, shrink = 0.8, label = "Temperature [°C]")

# set y-axis labels
axs[0].set_ylabel("Latitude [°N]")
axs[2].set_ylabel("Latitude [°N]")
# set x-axis labels
axs[2].set_xlabel("Longitude [°E]")
axs[3].set_xlabel("Longitude [°E]")


# save the figure to a png file (change the extension to save to other formats)
plt.savefig("seasonal_max_temp.png", dpi = 300)

plt.show()
```

**Caption**: Maximum 2-meter air temperature in each season of the year 2021.  The red star in (a) shows the location of Portland, OR.
