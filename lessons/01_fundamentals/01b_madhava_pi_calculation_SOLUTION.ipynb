This notebook is designed to be a live demonstration of Python fundamentals applied to calculating Ï€ using the [Madhava-Leibniz series](https://en.wikipedia.org/wiki/Approximations_of_%CF%80#Middle_Ages)

The approximation is simple:

$$ \pi \approx 4 \cdot \left(1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} \ldots -(-1)^n \frac{1}{2n - 1}  \right) $$

```python
""" This cell calculates pi via the Madhava-Leibniz series: https://en.wikipedia.org/wiki/Approximations_of_%CF%80#Middle_Ages """

# initialize pi to 0
pi = 0

# set the number of terms in the sum
N = 10000

# initialize an empty list store each term, so we can plot convergence later
pi_terms = []

# loop from 1 to N, adding terms to pi
# such that pi ~= 4*(1 - 1/3 + 1/5 - 1/7 + 1/9 ... -(-1)**n / (2n - 1))
for n in range(1,N):
    if n == 1:
        term = 1
    else:
        term = -((-1)**n) / (2*n - 1)
    pi = pi + term
    
    # store the term
    pi_terms.append(4*pi)

# multiply pi by four to get the final answer
pi = 4 * pi

# print the answer
print(f"pi = {pi:2.7f}")
```

```python
""" Check our answer against the built-in value of pi """
# import the numpy library; give it the shorter name 'np'
import math

# print the true value of pi
print(math.pi)
```

```python
""" Plot the convergence of the series. """

# import the plotting library; give it the shorter name 'plt'
import matplotlib.pyplot as plt

# plot the terms of the series
plt.plot(pi_terms)

# use a log scale on the x-axis
plt.xscale('log')

# plot a horizontal line at the true value of pi
plt.axhline(math.pi, color='gray', linestyle='--')

# set a useful title
plt.title("Convergence of the Madhava-Leibniz series for pi")

# draw the plot
plt.show()
```
